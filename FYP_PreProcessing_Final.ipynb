### 3.2.2 Data Merging

#### Step 1: Merge studentInfo and studentReg on 'code_module_presentation_student'

```python
merged_df = pd.merge(studentInfo, studentReg,
                     on=['code_module_presentation_student'],
                     how='inner')
print(merged_df.shape)
merged_df.head()
```

#### Step 2: Merge studentAss and assessment on 'id_assessment'

```python
studentAss_merge_assessments = pd.merge(studentAss, assessments,
                                        on=['id_assessment'],
                                        how='inner')
print(studentAss_merge_assessments.shape)
studentAss_merge_assessments.head()
```

Create new column - code_module_presentation_student

```python
studentAss_merge_assessments['code_module_presentation_student'] = studentAss_merge_assessments['code_module_presentation'] + '-' + studentAss_merge_assessments['id_student'].astype(str)
studentAss_merge_assessments
```

```python
studentAss_merge_assessments[studentAss_merge_assessments['score'].isna()]
```

```python
# Filter rows where 'score' is NaN and 'date_submitted' is greater than 'ass_cut_off_date'
filtered_df = studentAss_merge_assessments[(studentAss_merge_assessments['score'].isna()) & 
                                           (studentAss_merge_assessments['date_submitted'] > studentAss_merge_assessments['ass_cut_off_date'])]
filtered_df
```

```python
# Fill NaN values in 'score' where 'score' is NaN and 'date_submitted' is greater than 'ass_cut_off_date'
mask = (studentAss_merge_assessments['score'].isna()) & (studentAss_merge_assessments['date_submitted'] > studentAss_merge_assessments['ass_cut_off_date'])
studentAss_merge_assessments.loc[mask, 'score'] = 0
# Replace NaN with 0 for selected rows
```

```python
# Print the DataFrame after filling NaN values
print(studentAss_merge_assessments[studentAss_merge_assessments['score'].isna()].shape)
studentAss_merge_assessments[studentAss_merge_assessments['score'].isna()]
```

The remaining missing values in 'score' is due to the incomplete data entry, so all the remaining missing values will be filled with '0'

```python
studentAss_merge_assessments['score'].fillna(0, inplace=True)
studentAss_merge_assessments[studentAss_merge_assessments['score'].isna()]
```

Calculate Final Score for Each Student for Each Module

```python
# Separate regular assessment and exams
regular_assessments = studentAss_merge_assessments[studentAss_merge_assessments['weight'] != 100]
exams = studentAss_merge_assessments[studentAss_merge_assessments['weight'] == 100]
```

Regular Assessments

```python
# Calculate the weighted score for regular assessments
regular_assessments['weighted_score'] = (regular_assessments['score'] * regular_assessments['weight']) / 100

# Sum the weighted scores of regular assessments
regular_scores = regular_assessments.groupby(['code_module_presentation_student'])['weighted_score'].sum().reset_index()

regular_scores
```

Exam

```python
exam_scores = exams[['code_module_presentation_student', 'score']].copy()
exam_scores.rename(columns={'score': 'exam_score'}, inplace=True)

exam_scores
```

```python
# Merge the regular scores with the exam scores
final_scores = pd.merge(regular_scores, exam_scores, on=['code_module_presentation_student'], how='outer')
final_scores
```

```python
# Replace NaN in weighted_score and exam scores with 0 (in case a student didn't take the exam or the assessment)
final_scores['weighted_score'].fillna(0, inplace=True)
final_scores['exam_score'].fillna(0, inplace=True)
final_scores
```

```python
# Calculate the final score by adding regular scores and exam scores
final_scores['final_score'] = final_scores['weighted_score'] + final_scores['exam_score']
final_scores
```

```python
final_scores.describe()
```

```python
# Merge the regular scores with the exam scores
studentAss_merge_assessments = pd.merge(studentAss_merge_assessments, final_scores, 
                                        on=['code_module_presentation_student'], how='outer')
studentAss_merge_assessments
```

```python
# Group by 'code_module_presentation_student' and calculate the mean of 'final_score'
studAssessment = studentAss_merge_assessments.groupby('code_module_presentation_student', as_index=False)['final_score'].mean()

studAssessment
```

#### Step 3: Merge merged_df and studentAssessments on 'code_module_presentation_student'

```python
updated_df = pd.merge(merged_df, studAssessment,
                     on=['code_module_presentation_student'],
                     how='outer')
print(updated_df.shape)
updated_df.head()
```

#### Step 4: Merge updated_df and courses on 'code_module_presentation'

```python
updated_df = pd.merge(updated_df, courses,
                      on=['code_module_presentation'],
                      how='outer')
print(updated_df.shape)
updated_df.head()
```

#### Step 5: Merge studentVle and vle on 'code_module_presentation_site

```python
studentVle_merge_vle = pd.merge(studentVle, vle,
                     on=['code_module_presentation_site'],
                     how='outer')
print(studentVle_merge_vle.shape)
studentVle_merge_vle.head()
```

##### Calculate Number of Clicks Before and After Course Starts

Check Unique Values for 's_interaction_date'

```python
studentVle_merge_vle['s_interaction_date'].unique()
```

As there is negative and positive data in the column 's_interaction_date', negative values mean that the student is accessing the material before the course starts and positive values refer to students accessing to the material after the course starts.

So in this way, number of clicks before the module starts may be one of the factor which can help to determine whether the student is interested or not and will perform well or bad in the course.

So, by calculating the number of clicks before and after the course, ones have an interest on the course will be known.

##### Create New Column 'click_type'

```python
# Create a new column 'click_type' to show the interaction with vle is before course or after course
studentVle_merge_vle['click_type'] = ['before_course' if date < 0 else 'after_course' for date in studentVle_merge_vle['s_interaction_date']]
print(studentVle_merge_vle.shape)
studentVle_merge_vle.head()                          
```

##### Calculate the sum of 'sum_click' group by 'code_module_presentation_student' and 'click_type'

```python
grouped = studentVle_merge_vle.groupby(['code_module_presentation_student','click_type'])['sum_click'].sum().reset_index()
grouped
```

New columns are derived from 'click_type' and track the sum of clicks  for entries categorised as 'before_course' and 'after_course' in the 'click_type' respectively

```python
studentVle_merge_vle['before_course_click'] = np.where(studentVle_merge_vle['click_type'] == 'before_course', studentVle_merge_vle['sum_click'], 0)
studentVle_merge_vle['after_course_click'] = np.where(studentVle_merge_vle['click_type'] == 'after_course', studentVle_merge_vle['sum_click'], 0)

studentVle_merge_vle
```

```python
studentVle_merge_vle_group = studentVle_merge_vle.groupby(['code_module_presentation_student'], as_index=False)[['sum_click', 'before_course_click', 'after_course_click']].sum()
print(studentVle_merge_vle_group.shape)
studentVle_merge_vle_group
```

#### Step 6: Merge updated_df and studentVle_merge_vle_group on 'code_module_presentation_student'

```python
updated_df = pd.merge(updated_df, studentVle_merge_vle_group,
                     on=['code_module_presentation_student'],
                     how='outer')
print(updated_df.shape)
updated_df
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to check for outliers using IQR
def check_outliers_iqr(df):
    num_cols = df.select_dtypes(include=np.number).columns.tolist()
    
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Identify outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        num_outliers = outliers.shape[0]
        
        print(f'{col}\nSkew: {round(df[col].skew(), 2)}')
        print(f'Mean: {round(df[col].mean(), 2)}')
        print(f'Median: {round(df[col].median(), 2)}')
        print(f'Number of outliers: {num_outliers}')
        
        plt.figure(figsize=(15, 4))
        plt.subplot(1, 2, 1)
        df[col].hist(grid=False)
        plt.ylabel('count')
        plt.title(f'Histogram of {col}')
        
        plt.subplot(1, 2, 2)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        
        plt.show()
```

```python
print("--- Check for Outliers using IQR ---")
check_outliers_iqr(updated_df)
```

```python
def check_missing_values(df):
    # Checking for missing values
    missing_values = df.isnull().sum()

    # Calculating percentages
    percentage_missing = (missing_values / len(df)) * 100

    # Creating a dataframe with the missing values info
    missing_info = pd.DataFrame({'Missing Values': missing_values, 'Percentage Missing (%)': percentage_missing})

    # Sorting the dataframe by the number of missing values in descending order
    missing_info_sorted = missing_info.sort_values(by='Missing Values', ascending=False)

    return missing_info_sorted
```
